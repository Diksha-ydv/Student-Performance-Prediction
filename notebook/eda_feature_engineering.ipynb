### Creating two new features "Total score" and "Average score"

```python
df["total_score"] = df["math_score"]+df["reading_score"]+df["writing_score"]
df["average"] = df["total_score"]/3
df.head()
```

### Getting number of students scoring full marks 

```python
maths = df[df["math_score"]==100]["average"].count()
reading = df[df["reading_score"]==100]["average"].count()
writing = df[df["writing_score"]==100]["average"].count()

print(f"Number of students getting full marks in maths are : {maths}")
print(f"Number of students getting full marks in reading are : {reading}")
print(f"Number of students getting full marks in writing are : {writing}")
```

### Observation: Best performance is in reading

### Getting number of students performing poorly

```python
maths = df[df["math_score"]<=20]["average"].count()
reading = df[df["reading_score"]<=20]["average"].count()
writing = df[df["writing_score"]<=20]["average"].count()

print(f"Number of students getting less marks in maths are : {maths}")
print(f"Number of students getting less marks in reading are : {reading}")
print(f"Number of students getting less marks in writing are : {writing}")
```

### Observation: Worst performance is in maths
